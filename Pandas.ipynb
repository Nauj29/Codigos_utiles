```python
#Importamos libreria Pandas
import pandas as pd
import numpy as np
from osgeo import gdal
import glob
import os
from pathlib import Path
```

```python
#definidimos la carpeta en que estan los datos
os.chdir(r"D:\UNAM\Doctorado\SIG\Clima\Estaciones_climaticas")
```

```python
inputFiles = glob.glob("*.txt")
inputFiles
```

```python
for i in inputFiles[:1]:
    d=Path(i)
    f=d.stem
    df=pd.read_csv(i,skiprows=18,sep = "\s+", skipfooter=1, engine='python')
    df['FECHA'] = pd.to_datetime(df['FECHA'],dayfirst=True, format='%d/%m/%Y')
    df.index=df["FECHA"]
    df.drop("FECHA",axis=1, inplace=True)
    df.columns=["P(mm)","E(mm)","Tmx(°C)","Tmn(°C)"]
    idx=pd.date_range(df.index.min(),df.index.max())
    df.index = pd.DatetimeIndex(df.index)
    
    z=df.reindex(idx, fill_value='Nulo')
    z.index.name="Fecha"
    z=z.replace('Nulo','NaN')
    
    d=z["P(mm)"]
    c=z[["Tmn(°C)","Tmx(°C)"]]
    
    d.to_csv(r"D:\UNAM\Doctorado\SIG\Clima\Precipitacion"+f+"P.txt", sep = " ")
    c.to_csv(r"D:\UNAM\Doctorado\SIG\Clima\Temperatura"+f+"T.txt", sep = " ")





```
